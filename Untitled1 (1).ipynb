```python
import pandas as pd 
import numpy as np
import ast
```

```python
movies = pd.read_csv('movies_10k.csv',index_col=0)
```

```python
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
def convert(object):
    list=[]
    for i in ast.literal_eval(object):
        list.append(i['name'])
    return list
```

```python
def convert3(obj):
    L = []
    c=0
    for i in ast.literal_eval(obj):
        if(c!=3):
            L.append(i['name'])
            c+=1
        else:
            break
    return L
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies['crew'] = movies['crew'].apply(convert)
```

```python
movies.head()
```

```python
movies = movies[movies['genre_name'] != '[]']
```

```python
movies.shape
```

```python
movies['genre_name'] = movies['genre_name'].apply(lambda x: x.strip("[]"))
```

```python
movies.head()
```

```python
movies['genre_name'] = movies['genre_name'].apply(lambda x: x.replace(",", ''))
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genre_name'] = movies['genre_name'].apply(lambda x: x.split())
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genre_name']=movies['genre_name'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview']+ movies['genre_name'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['id','title','tags']]
new_df.shape
```

```python
poster_df = pd.read_csv('poster.csv',index_col=0)
poster_df
```

```python
new_df = new_df.merge(poster_df,on='id')
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df
```

```python
new_df.to_csv('final_df.csv')
```

```python
new_df.shape
```

```python
new_df['tags'] =  new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer(max_features =5000, stop_words='english')
```

```python
vectors=v.fit_transform(new_df['tags']).toarray()
```

```python
vectors, vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    #find the index of the movies
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    #to fetch movies from indeces
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
